<h1>Matplotlib Part 1</h1>

```python
import matplotlib.pyplot as plt
```

```python
# this allows you to see your figures in Jupyter Notebook
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL METHOD
plt.plot(x,y) #,'r--')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
# if not in jupyter notebook you will 
# need plt.show() to show the plot
```

```python
plt.subplot(1,2,1) # 3rd argument refers to the subplot. This is the first hence it's 1
plt.plot(x,y,'r')

plt.subplot(1,2,2) # Again, 3rd argument refers to the subplot. This is the second hence it's 2
plt.plot(y,x,'b')
```

```python
# OBJECT METHOD
fig = plt.figure()

# the list in add_axes takes in 4 arguments:
# left, bottom, width and height ranging from 0 -1
axes = fig.add_axes([0.1,0.1,0.8,0.8]) 

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.5,0.18,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

<h1>Matplotlib Part 2</h1>

```python
# fig,axes is an example of tuple unpacking
fig,axes = plt.subplots(nrows=1,ncols=2)

# axes.plot(x,y)

# cleans up an overlapping lines and numbers
# plt.tight_layout() 

# for current_ax in axes:
#     current_ax.plot(x,y)

# index calls through the axes array
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

<h2>Figure Size and DPI</h2>

```python
# figsize argument takes in an argument that is a tuple of the width and height
# dpi is another possible argument and is the dots or pixels per inch
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

<h2>Matplotlib Part 3</h2>

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=2,alpha=0.5,linestyle='-',marker='o',markersize=20,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green') 
# linewidth default is 1
# you can also just type lw for linewidth (i.e. lw=3)
# you enter color strings like, blue, green, orange, purple
# you can also use rgb hex codes (i.e. '#FF8C00')
# linestyle reference: https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html
```
