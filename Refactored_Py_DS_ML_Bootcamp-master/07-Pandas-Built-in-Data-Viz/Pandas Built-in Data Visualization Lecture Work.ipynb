<h1>Pandas Built-in Data Visualization</h1>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
```

```python
df1 = pd.read_csv('df1',index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

```python
df1['A'].hist()
```

```python
df1['A'].hist(bins=30)
```

```python
df1['A'].plot(kind='hist')
```

```python
df1['A'].plot.hist() # preferred method compared to the 2 above
```

```python
df2.head()
```

```python
df2.plot.area()
```

```python
df2.plot.area(alpha=0.4)
```

```python
df2.plot.bar(stacked=True)
```

```python
df1.plot.line(y='B')
```

```python
plt.style.use('dark_background')
df1.plot.line(y='B',figsize=(12,3),lw=1)
```

```python
plt.style.use('fivethirtyeight')
df1.plot.scatter(x='A',y='B',c='C')
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*100,edgecolors='black') # edgecolors adds a border to the bubbles
```

```python
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

```python
df.head()
```

```python
df.plot.hexbin(x='a',y='b') # hexbin is used for bivariate datasets (having 2 columns)
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25)
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='coolwarm')
```
